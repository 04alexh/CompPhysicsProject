```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("geodesic_output.npz", allow_pickle=True)

#extracting
t = data["t"]
y = data["y"]
coords = data["coords"]
```

```python
#practice plotitng functions 

#sol.y[0:4] → coordinates (t, x1, x2, x3)
#sol.y[4:8] → velocities

coord_map = {coords[i]: i for i in range(len(coords))}

r = y[coord_map["r"]]
phi = y[coord_map["phi"]]

plt.plot(phi, r)
plt.xlabel("phi")
plt.ylabel("r")
plt.title("Orbit")
plt.show()
```
