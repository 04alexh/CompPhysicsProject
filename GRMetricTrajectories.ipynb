### Simple Christoffel Symbol Calc

this was the christoffel symbol thing i made

```python
import numpy as np
import sympy as sp

t , r , theta , phi= sp.symbols('t r theta phi')
M = 5
g = sp.Matrix([[-(1 - 2*M/r) , 0 , 0 , 0] , [0 , 1/(1 - 2*M/r) , 0, 0] , [0 , 0 , r**2 , 0] , [0 , 0 , 0 , r**2 * sp.sin(theta)**2]])
print(g)

indices = {0:t , 1:r , 2:theta , 3:phi}
print(indices[0])
ginv = g.inv()
print(ginv)

christoffel = sp.MutableDenseNDimArray.zeros(4,4,4)
for mu in range(0 , 4):
    for nu in range(0 , 4):
        for rho in range(0 , 4):
            expr = 0
            for sig in range(0 , 4):
                expr += .5 * ginv[mu , sig] * (
                    sp.diff(g[rho , sig] , indices[nu]) +
                    sp.diff(g[nu , sig] , indices[rho]) -
                    sp.diff(g[nu , rho] , indices[sig])
                )
            christoffel[mu , nu , rho] = expr

print(christoffel)
#
# dlam = .01
# tlam = 1000
# nlam = int(tlam / dlam)
#
# lam = np.zeros(nlam)
#
# for l in range(0 , nlam - 1):


```
